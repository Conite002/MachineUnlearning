```python
import sys
import os
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
import warnings
import time
# Append the system path to include the parent directory for custom modules
sys.path.append('../')

# Import necessary modules from the custom packages
from Unlearner.DPLRUnlearner import DPLRUnlearner
from Applications.FeatureUnlearning.LinearEnsembleExperiments import split_train_data, create_models
from Applications.FeatureUnlearning.DataLoader import DataLoader
from Applications.FeatureUnlearning.LRExperiments import get_average_gradient_residual, scatter_experiments, find_most_relevant_indices
```

```python
# Suppress warnings for cleaner output
warnings.filterwarnings("ignore")

# List of datasets to be used for the comparative study
datasets_name = ['Adult', 'Diabetis', 'Drebin', 'Enron']

# Unlearning parameters
lambda_ = 1.0  # Regularization strength
sigma = 0.0  # Noise to compute gradient residual
combination_lengths = [10, 50, 100]  # Number of features to remove
n_combinations = 10  # Number of combinations of features to remove
unlearning_rate = 0.0001  # First Order update unlearning rate
unlearning_rate_ft = unlearning_rate  # Learning rate for fine-tuning
iter_steps = 10  # Number of steps for iterative unlearning
remove = True  # Flag to actually remove the features

csv_file = 'unlearning_performance_metrics.csv'

# Function to perform unlearning and save results
def perform_unlearning(dataset_name, normalize=True):
    # Check if the dataset is supported
    supported_datasets = ['Adult', 'Diabetis', 'Enron', 'Drebin']
    if dataset_name not in supported_datasets:
        print(f"Dataset {dataset_name} is not supported. Skipping...")
        return

    # Load the training and test data
    loader = DataLoader(dataset_name, normalize)
    train_data, test_data, voc = (loader.x_train, loader.y_train), (loader.x_test, loader.y_test), loader.voc
    category_to_idx_dict = loader.category_to_idx_dict

    # Create folders for results
    res_save_folder = f'Results_{dataset_name}'
    model_save_folder = f'Models_{dataset_name}'
    if not os.path.isdir(res_save_folder):
        os.makedirs(res_save_folder)

    # Identify relevant features to be deleted
    relevant_features = loader.relevant_features
    relevant_indices = [voc[f] for f in relevant_features]
    indices_to_delete = relevant_indices

    print(f"Dataset: {dataset_name}")
    print(f"Relevant features: {', '.join(relevant_features[:10])}...")
    print(f"Number of relevant features: {len(relevant_features)}")

    # Check if combination lengths exceed the number of available features
    actual_combination_lengths = [cl for cl in combination_lengths if cl <= len(indices_to_delete)]

    if not actual_combination_lengths:
        print(f"Not enough features to remove for dataset {dataset_name}. Skipping...")
        return

    start_time = time.time()
    # Perform unlearning and get gradient residuals
    get_average_gradient_residual(train_data, test_data, voc, lambda_, sigma, indices_to_delete,
                                  actual_combination_lengths, n_combinations, unlearning_rate, unlearning_rate_ft,
                                  iter_steps, category_to_idx_dict, remove=remove, save_path=res_save_folder)

    gradient_residual_time = time.time() - start_time
    print(f"Time taken for get_average_gradient_residual: {gradient_residual_time:.2f} seconds")

    # Additional fidelity evaluation
    combination_length = 20
    most_important_size = 300
    top_indices, _ = find_most_relevant_indices(train_data, test_data, voc, top_n=most_important_size)
    indices_to_delete = np.intersect1d(top_indices, relevant_indices)

    # Check if combination length for scatter experiments exceeds the number of available features
    if combination_length > len(indices_to_delete):
        print(f"Not enough features to remove for scatter experiments in dataset {dataset_name}. Skipping scatter experiments...")
        return
    
    start_time = time.time()
    scatter_experiments(train_data, test_data, voc, lambda_, sigma, indices_to_delete, combination_length,
                        unlearning_rate, unlearning_rate_ft, n_combinations, category_to_idx_dict, remove)
    scatter_experiments_time = time.time() - start_time
    print(f"Time taken for scatter_experiments: {scatter_experiments_time:.2f} seconds")

    # Save the times to the CSV file
    performance_metrics = pd.DataFrame({
        'Dataset': [dataset_name],
        'Gradient_Residual_Time': [gradient_residual_time],
        'Scatter_Experiments_Time': [scatter_experiments_time]
    })
    if os.path.isfile(csv_file):
        existing_data = pd.read_csv(csv_file)
        combined_data = pd.concat([existing_data, performance_metrics], ignore_index=True)
        combined_data.to_csv(csv_file, index=False)
    else:
        performance_metrics.to_csv(csv_file, index=False)

# Iterate through each dataset and perform unlearning
for dataset_name in datasets_name:
    perform_unlearning(dataset_name)
```

```python
import pandas as pd

# List of paths to the CSV files
folders = ['Results_Adult/fidelity_results.csv', 'Results_Drebin/fidelity_results.csv', 'Results_Enron/fidelity_results.csv']

# Initialize an empty list to store individual DataFrames
dfs = []

# Iterate over the file paths and read each CSV into a DataFrame
for f in folders:
    df = pd.read_csv(f)
    
    # Extract dataset name from the file path
    dataset_name = f.split('_')[1].split('/')[0]
    
    # Update the dataset name with the combination type
    df['Datasets'] = dataset_name + '_' + df['x'].astype(str)
    
    # Append the DataFrame to the list
    dfs.append(df)

# Combine all the DataFrames into a single DataFrame
combined_df = pd.concat(dfs, ignore_index=True)

# Display the combined DataFrame
print(combined_df)
```

```python
# Regrouper par 'x'
grouped_df = combined_df.groupby('x')

# Créer les sous-graphes
fig, axs = plt.subplots(1, len(grouped_df), figsize=(20, 8), sharey=True)

# Couleurs des barres pour chaque méthode
colors = ['b', 'g', 'r', 'c', 'm', 'y']

# Tracer les barres pour chaque groupe de 'x'
for (x_value, group), ax in zip(grouped_df, axs):
    group = group.set_index('Datasets')
    group.drop('x', axis=1, inplace=True)
    group.plot(kind='bar', ax=ax, color=colors, legend=False)
    ax.set_title(f'x = {x_value}')
    ax.set_xlabel('Datasets')
    ax.set_ylabel('Accuracy')

# Ajouter une légende globale
handles, labels = axs[-1].get_legend_handles_labels()
fig.legend(handles, labels, title='Unlearning Methods', bbox_to_anchor=(1.05, 0.5), loc='center left')

plt.tight_layout()
plt.show()
```
